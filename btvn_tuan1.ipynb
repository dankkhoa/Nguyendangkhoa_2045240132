```python
# Bài 1:
def bai1():
    # Nhập mảng a
    a = list(map(int, input("Nhập mảng a (các số cách nhau): ").split()))
    # Nhập mảng b
    b = list(map(int, input("Nhập mảng b (các số cách nhau): ").split()))
    result = []
    min_len = min(len(a), len(b))
    for i in range(min_len):
        result.append(min(a[i], b[i])) # Changed to min instead of sum
    # Thêm phần tử còn lại
    if len(a) > min_len:
        result.extend(a[min_len:])
    if len(b) > min_len:
        result.extend(b[min_len:])
    print("Mảng kết quả:", result)
bai1()
```

```python
# Define a sample square matrix
a = [
    [1, 2, 3],
    [4, -5, 6],
    [7, 8, 9]
]

print("Original matrix:", a)

# a. Tổng tam giác trên của đường chéo phụ (kể cả đường chéo phụ)
def sum_triangle_above_secondary_diag(matrix):
    n = len(matrix)
    total = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:
                total += matrix[i][j]
    return total


# b. Chuyển số âm thành trị tuyệt đối
def abs_negative(matrix):
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j] < 0:
                matrix[i][j] = abs(matrix[i][j])


# c. Thay số chẵn bằng x
def replace_even(matrix, x):
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j] % 2 == 0:
                matrix[i][j] = x


# d. Kiểm tra toàn chẵn
def all_even(matrix):
    for row in matrix:
        for val in row:
            if val % 2 != 0:
                return False
    return True


# e. Kiểm tra đối xứng
def is_symmetric(matrix):
    n = len(matrix)
    for i in range(n):
        for j in range(n):
            if matrix[i][j] != matrix[j][i]:
                return False
    return True


# f. Đường chéo chính tăng dần
def main_diag_increasing(matrix):
    n = len(matrix)
    for i in range(n - 1):
        if matrix[i][i] >= matrix[i + 1][i + 1]:
            return False
    return True


# g. In tam giác dưới của đường chéo phụ
def print_triangle_below_secondary_diag(matrix):
    n = len(matrix)
    for i in range(n):
        for j in range(n):
            if i + j >= n - 1:
                print(matrix[i][j], end=" ")
        print()


# h. Đường chéo phụ giảm dần
def secondary_diag_decreasing(matrix):
    n = len(matrix)
    for i in range(n - 1):
        if matrix[i][n - 1 - i] <= matrix[i + 1][n - 2 - i]:
            return False
    return True


```
